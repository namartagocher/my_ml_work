```python
def classifier_visual(m,x,y,ax=None,cmap='rainbow'):
    ax=plt.gca()
    ax.scatter(x[:,0],x[:,1],c=y,s=50,cmap='rainbow',clim=(y.min(),y.max()),zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim=ax.get_xlim()
    ylim=ax.get_ylim()
    m.fit(x,y)
    xx,yy=np.meshgrid(np.linspace(*xlim,num=200),np.linspace(*ylim,num=200))
    z=m.predict(np.c_[xx.ravel(),yy.ravel()]).reshape(xx.shape)
    n_class=len(np.unique(y))
    cont=ax.contourf(xx,yy,z,alpha=0.3,levels=np.arange(n_class+1)-0.5,cmap='rainbow',clim=(y.min(),y.max()),zorder=1)
    ax.set(xlim=xlim,ylim=ylim)
```

```python
classifier_visual(DecisionTreeClassifier(),x,y)
```

### noise

```python
rang=np.random.RandomState(42)
x=10*rang.rand(200)
```

```python
len(x)
```

```python
x
```

```python
def method(x,sigma=0.3):
    fast_oscillation=np.sin(x*5)
    slow_oscillation=np.sin(0.5*x)
    noise=sigma*rang.randn(len(x))
    return fast_oscillation+slow_oscillation+noise
    
```

```python
y=method(x)
plt.errorbar(x,y,0.3,fmt='o')
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rm=RandomForestRegressor(200)
rm.fit(x[:,None],y)
```

```python
x_fit=np.linspace(0,10,1000)
y_fit=rm.predict(x_fit[:,None])
y_true=method(x_fit,0)
plt.errorbar(x,y,0.3,alpha=0.5,fmt='o')
plt.plot(x_fit,y_fit,'-r')
plt.plot(x_fit,y_true,'-g',alpha=0.5)#difference is noise data
```

```python
from sklearn.datasets import load_digits
```

```python
digit=load_digits()
digit.keys()
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(6,6))
for i in range(64):
    ax=fig.add_subplot(8,8,i+1,xticks=[],yticks=[])
    ax.imshow(digit.images[i],cmap=plt.cm.binary,interpolation='nearest')
    ax.text(0,7,str(digit.target[i]))
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test=train_test_split(digit.data,digit.target,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=1000) 
```

```python
model.fit(x_train,y_train)
y_pre=model.predict(x_test)
```

```python
from sklearn import metrics
print(metrics.classification_report(y_pre,y_test))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
mat=confusion_matrix(y_test,y_pre)
sns.heatmap(mat.T,square=True,annot=True,cbar=False)
plt.xlabel('true value')
plt.ylabel("predicted value")
```
